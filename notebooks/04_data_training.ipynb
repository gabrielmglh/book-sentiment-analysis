```python
# Loading the model

reviews_model = joblib.load('data/model/best_model.joblib')
```

## Testing the model

Now let us write some reviews to test wether the model was well trained.

```python
reviews = [
    "I absolutely loved this book! The story was gripping from start to finish.",
    "One of the worst books I've ever read. The plot makes no sense.",
    "It was an okay read, nothing special but not terrible either.",
    "Fantastic writing style and deeply engaging characters. Highly recommended!",
    "I struggled to finish it. Very boring and repetitive.",
    "A masterpiece. The author did an amazing job building the world and characters.",
    "Not my cup of tea. I found it dull and predictable.",
    "Really enjoyable book with a strong ending. I would read it again.",
    "Poorly written and hard to follow. I expected much more.",
    "An average book with some good moments but also some weak parts."
]
```

Now we create a function based on this model that takes the review and returns its sentiments

```python
def review_sentiment(reviews):
    
    """
    Takes a list of reviews and return the sentiment of the given review.
    """

    predictions = reviews_model.predict(reviews)
    
  
    sentiments = ['Negative' if p == 0 else 'Positive' for p in predictions]
    
    
    for review, sentiment in zip(reviews, sentiments):
        print(f"\nReview: '{review}'\nPredicted sentiment: {sentiment}\n---")
```

Now we test it

```python
review_sentiment(reviews)
```

Thus we note that out of our 10 reviews provided, just the last one is not quite right the prediction. But in the overall, we have a well-trained model with a good accuracy.
