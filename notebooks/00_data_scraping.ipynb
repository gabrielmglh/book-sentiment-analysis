```python
import os
from pathlib import Path

# Defines a function to find the project root directory by looking for a specific marker (default is 'data')
def find_project_root(start=Path.cwd(), marker='data'): 
    current = start
    while current != current.parent:
        if (current / marker).exists():
            return current
        current = current.parent
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root() 
os.chdir(PROJECT_ROOT) # go to the root of the project 
```

```python
# --- Locate kaggle.json relative to this notebook ---
KAGGLE_JSON   = PROJECT_ROOT / 'kaggle' / "kaggle.json"
KAGGLE_DIR    = Path.home() / ".kaggle"
DATA_DIR      = PROJECT_ROOT / "data" / "dataframes" / "bronze"

assert KAGGLE_JSON.exists(), (
    f"kaggle.json not found at {KAGGLE_JSON}. "
    "Place it in the repository root before running this notebook."
)

# Copy credentials to ~/.kaggle/ (required by the kaggle CLI)
KAGGLE_DIR.mkdir(exist_ok=True)
import shutil
shutil.copy(KAGGLE_JSON, KAGGLE_DIR / "kaggle.json")
os.chmod(KAGGLE_DIR / "kaggle.json", 0o600)   # Kaggle requires strict permissions

DATA_DIR.mkdir(exist_ok=True)
print("Setup complete.")
```

```python
# --- Download & unzip the dataset ---
# Dataset: https://www.kaggle.com/datasets/meetnagadia/amazon-kindle-book-review-for-sentiment-analysis
!kaggle datasets download -d meetnagadia/amazon-kindle-book-review-for-sentiment-analysis \
    --path "{DATA_DIR}" --unzip

print("\nFiles in data/:")
for f in sorted(DATA_DIR.iterdir()):
    print(" ", f.name)
```
