```python
import os
import zipfile
from pathlib import Path
from kaggle.api.kaggle_api_extended import KaggleApi
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Resolve project root dynamically
BASE_DIR = Path(__file__).resolve().parent.parent

# Read env variables (relative paths → absolute at runtime)
KAGGLE_DIR = BASE_DIR / os.getenv("KAGGLE_CONFIG_DIR", "kaggle")
DATA_DIR = BASE_DIR / os.getenv("DATA_DIR", "data")

def setup_environment():
    # Ensure directories exist
    KAGGLE_DIR.mkdir(parents=True, exist_ok=True)
    DATA_DIR.mkdir(parents=True, exist_ok=True)

    # Set environment variable for Kaggle API
    os.environ["KAGGLE_CONFIG_DIR"] = str(KAGGLE_DIR)

def download_dataset(dataset: str):
    api = KaggleApi()
    api.authenticate()

    api.dataset_download_files(dataset, path=DATA_DIR, unzip=True)

    print(f"Dataset '{dataset}' downloaded to {DATA_DIR}")

if __name__ == "__main__":
    setup_environment()
    download_dataset("zynicide/wine-reviews")
```
